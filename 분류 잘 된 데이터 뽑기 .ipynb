```python
import os

data_path='./data'
files=os.listdir(data_path)
data={"text":[], "label":[]}

for file in files:
    file_path=os.path.join(data_path, file)
    with open(file_path, 'r', encoding='utf-8') as f:
        lines=f.readlines()
        for line in lines:
            data["text"].append(line.split('\t')[0])
            data["label"].append(line.split('\t')[1].strip())

print(data["text"][:10])
print(data["label"][:10])
```
