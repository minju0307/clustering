```python
import os

data_path='./data'
files=os.listdir(data_path)
train_data={"text":[], "label":[]}
dev_data={"text":[], "label":[]}
test_data={"text":[], "label":[]}

for file in files:
    file_path=os.path.join(data_path, file)
    with open(file_path, 'r', encoding='utf-8') as f:
        lines=f.readlines()
        for line in lines:
            if 'train' in file:
                train_data["text"].append(line.split('\t')[0])
                train_data["label"].append(line.split('\t')[1].strip())
            if 'dev' in file:
                dev_data["text"].append(line.split('\t')[0])
                dev_data["label"].append(line.split('\t')[1].strip())
            if 'test' in file:
                test_data["text"].append(line.split('\t')[0])
                test_data["label"].append(line.split('\t')[1].strip())
            

print(train_data["text"][:10])
print(train_data["label"][:10])
```

```python
train_data_tuple=[]
for i, text in enumerate(train_data["text"]):
    train_data_tuple.append((text, train_data["label"][i]))
```
